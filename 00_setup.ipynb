Wusstest Du, dass in HSV nun robuste (d.h. von Beleuchtung unabhängige) Farbhistogramme möglich sind? Erstelle eines mit Matplotlib!

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from skimage import io, data, color

img_hsv = color.rgb2hsv(image)
a = np.histogram(img_hsv, bins=10, range=None, normed=False, weights=None, density=None)
plt.hist(a)
plt.show()
```
